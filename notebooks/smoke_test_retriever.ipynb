```python
from sentence_transformers import SentenceTransformer
from sentence_transformers.util import cos_sim
```

```python
model = SentenceTransformer("C:/Users/sofiy/PycharmProjects/nlp-project/models/models/retriever")
print("Model loaded")
```

```python
query = "query: Jak działa system zdrowotny w Polsce?"

passages = [
    "passage: W Polsce obowiązuje powszechny system opieki zdrowotnej.",
    "passage: Najwyższy szczyt Tatr to Rysy.",
    "passage: Państwowy system zdrowia finansowany jest z podatków."
]
```

```python
q_emb = model.encode([query], normalize_embeddings=True)
p_embs = model.encode(passages, normalize_embeddings=True)

scores = cos_sim(q_emb, p_embs)[0]
```

```python
for passage, score in zip(passages, scores):
    print(f"{score:.4f} ← {passage}")
```
